```python
# Libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generating data for sin(x) curve
start = int(input("Enter starting point : "))
end = int(input("Enter ending point : "))
numPoints = int(input("Enter number of points : "))

x = np.linspace(start, end, num = numPoints)
y = np.sin(x)
```

```python
# Plotting sin(x) curve
plt.figure(figsize = (10,5))
plt.plot(x, y)
plt.title("Sin(x) curve")
plt.xlabel("Range %d-%d"%(start, end))
plt.ylabel("Values")
```

```python
# Getting gauss error and adding it to y
gaussian_error = 0.01*np.random.normal(0, 10, numPoints)
gaussian_error.shape

y_gaussian_error = y + gaussian_error

plt.figure(figsize = (10,5))
plt.plot(x, y_gaussian_error, "r-")
plt.title("Sin(x) curve with error")
plt.xlabel("Range %d-%d"%(start, end))
plt.ylabel("Values")
```

```python
# Order of equation = 0
x_temp = list()
for i in x:
    x_temp.append([i**0])
x_temp = np.array(x_temp)

x_temp.shape

# Transpose and inverse
xt = np.matrix.transpose(x_temp)
xtx = xt.dot(x_temp)
inverse_xtx = np.linalg.inv(xtx)

xty = xt.dot(y_gaussian_error)
w = inverse_xtx.dot(xty)

y_temp = list()
x_temp = list()
for i in x:
    y_temp.append([w[0]* i**0])
    x_temp.append(i)

x_temp = np.array(x_temp)
y_temp = np.array(y_temp)

plt.figure(figsize = (10,5))
plt.plot(x_temp, y_temp, "r-")
plt.scatter(x, y_gaussian_error)
plt.title("Sin(x) curve with error")
plt.xlabel("Range %d-%d"%(start, end))
plt.ylabel("Values")
```
