```python
# Libraries included
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math 
```

```python
# You can choose any value
numPoints = 100
```

```python
# Select your starting and ending points
start = int(input("Enter starting point of range : "))
end = int(input("Enter starting point of range : "))
```

```python
# This function distributes range into various data point equidistant from each other (adjacent element) on a number line  
# to divides whole range into numPoints parts 
# If numpoint is equal to 10 then it will divide it into 10 equal points
x = np.linspace(start, end, num = numPoints)
```

```python
x.shape
```

```python
# Getting y for respective x
y = np.sin(2 * np.pi * x)
```

```python
y.shape
```

```python
# Plotting sinx Curve without noise
```

```python
plt.figure(figsize = (10,5))

plt.plot(x, y)
plt.title("Sine Curve without noise")
plt.xlabel("Values of x in range(%d, %d)"%(start, end))
plt.ylabel("Sin value ")
```

```python
# Noise function 
# Gaussian error - Took it straight from the internet
eta = .4*np.random.normal(start,end,numPoints)
eta.shape
```

```python
# Noise curve
# Gaussian noise generated curve i.e. eta curve wtr to x
plt.figure(figsize = (10,5))

plt.plot(x, eta,  "r-")
plt.scatter(x, eta)
plt.title("Gaussian noise")
plt.xlabel("Range(%d, %d)"%(start, end))
plt.ylabel("Gaussian Error ")
```

```python
# Adding noise to sine wave
y_noise = y + eta
y_noise.shape
```

```python
# After adding noise to original sine curve
# Noised Sine curve
plt.figure(figsize = (10,5))

x_eta = np.linspace(start, end, 10)
eta = 0.2*np.random.normal(start,end,10)
y_eta = np.sin(2*np.pi*x_eta)+ eta

plt.plot(x,y, "r-")
plt.scatter(x_eta, y_eta)
plt.title("Gaussian noise")
plt.xlabel("Range(%d, %d)"%(start, end))
plt.ylabel("Gaussian Error ")
```

```python
# Now fitting polynomial
# 
```

```python
# Order here represents the power of x
order = 0
poly = np.polyfit(x_eta, y_eta, order)

polyArray = np.poly1d(poly)
y_poly = polyArray(x_eta)

plt.figure(figsize = (10,5))
plt.scatter(x_eta, y_eta)
plt.plot(x_eta, y_poly, "r-")
```
